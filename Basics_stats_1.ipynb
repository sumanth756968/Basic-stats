```python
import numpy as np
import pandas as pd
df=pd.read_csv("sales_data_with_discounts.csv")
df.head()
```

The dataset is loaded using the Pandas library into a DataFrame named df. The head() function is used to display the first five rows of the dataset.

```python
df.describe()
```

The describe() function is used to generate summary statistics for the numerical columns in the dataset, including count, mean, standard deviation, minimum, maximum, and quartile values.

```python
df.info()
```

The info() function is used to display the dataset's structure, including the number of rows, columns, data types, non-null values, and memory usage.

```python
numerical_columns = df.select_dtypes(include=np.number).columns
categorical_columns = df.select_dtypes(include='object').columns

print("Numerical Columns:")
print(numerical_columns)

print("\nCategorical Columns:")
print(categorical_columns)
```

The select_dtypes() function is used to separate the numerical and categorical columns in the dataset. This helps apply appropriate analysis and preprocessing techniques, as numerical and categorical data require different methods. The names of these columns are then displayed.

```python
print("Mean")
print(df[numerical_columns].mean())
print("Median")
print(df[numerical_columns].median())
print("Mode")
print(df[numerical_columns].mode().iloc[0])
print("Standard Deviation")
print(df[numerical_columns].std())
```

The mean, median, mode, and standard deviation are calculated for the numerical columns to summarize the data. These statistics help understand the central tendency and variability of the dataset.

**Interpretation**
The descriptive statistics provide a summary of the numerical data in the dataset. The mean shows the average value, the median represents the middle value, and the mode indicates the most frequently occurring value. The standard deviation measures how much the data varies from the mean. Together, these measures help understand the central tendency and variability of the dataset, making it easier to identify patterns and prepare the data for further analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns
for column in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[column], bins=20, kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
```

The above code generates histogram plots for all numerical columns to visualize their data distributions.

```python
for column in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

The above code generates Boxplot for all numerical columns to visualize their data distributions.

```python
for column in categorical_columns:
    plt.figure(figsize=(8,4))
    sns.countplot(x=df[column])
    plt.title(f'Bar Chart of {column}')
    plt.xticks(rotation=45)
    plt.show()
```

The above code generates Bar Chart for all numerical columns to visualize their data distributions.

**Interpretation**
The histogram shows the distribution and frequency of values in each numerical column, helping identify the shape and spread of the data. The box plot summarizes the data distribution by displaying the median, quartiles, and possible outliers, making it useful for detecting variability and extreme values. The bar chart represents the frequency of each category in categorical columns, allowing easy comparison of different categories and helping identify the most and least common values in the dataset.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

standardized_data = scaler.fit_transform(df[numerical_columns])

standardized_df = pd.DataFrame(
    standardized_data,
    columns=numerical_columns
)

print(standardized_df.head())
```

The numerical features were standardized using StandardScaler, which transforms each feature to have a mean of 0 and a standard deviation of 1. This ensures that all numerical variables are on the same scale, preventing features with larger values from dominating the analysis. Standardization improves the performance of many machine learning algorithms and makes the dataset suitable for further modeling.

```python
print(df[numerical_columns].head())
```

```python
print(standardized_df.head())
```

It shows the difference between the data before standardization and after standardization.

```python
for col in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Before Standardization - {col}')
    plt.show()
```

```python
for col in standardized_df.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(standardized_df[col], kde=True)
    plt.title(f'After Standardization - {col}')
    plt.show()
```

It shows difference in Histogram plot

```python
print(categorical_columns)
```

```python
dummy_df = pd.get_dummies(df, columns=categorical_columns)

print(dummy_df.head())
```

The above code performs one-hot encoding on all categorical columns using pd.get_dummies(). It converts each category into separate binary (0 or 1) columns, making the dataset suitable for machine learning algorithms. The head() function displays the first five rows of the transformed dataset to verify the encoded output.

```python
print(dummy_df.columns)
```

```python
dummy_df.to_csv("Sales_Data_Dummy.csv", index=False)
```

The above code saves the dataset with one-hot encoded (dummy) variables to a CSV file named Sales_Data_Dummy.csv. The parameter index=False ensures that the row index is not included in the saved file, producing a clean dataset for further analysis or machine learning.
